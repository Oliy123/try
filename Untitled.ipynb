```python
from libpysal.weights import Queen
import libpysal
from libpysal.weights import Queen, Rook, KNN, weights, lag_spatial
```

## patial weights matrix. 
spatial weights matrix is the way geographical space is formally encoded into a numerical form. These are structured sets of numbers that formalize geographical relationships between the observations in a dataset.



```python
w = KNN.from_dataframe(br, k=5)
w
```

```python
w.transform = "R"
```

```python
br['totCases'] = br["totCases"].astype(int)
```

## Spatial lag
 Once we have the data and the spatial weights matrix ready, we can start by computing the spatial lag of Number Covid 19 case in every county. spatial lag is the product of the spatial weights matrix and a given variable and that, if W is row-standardized, the result amounts to the average value of the variable in the neighborhood of each observation  

```python
y = np.log(br['totCases'])
y
```

```python
br['w_totcases'] = weights.lag_spatial(w, y)
```

```python
br.head()
```

```python
br['cases_Leave_std'] = (
    br['totCases'] - br['totCases'].mean()
) / br['totCases'].std()
```

```python
br['cases_Leave_std'] = weights.lag_spatial(w, br['cases_Leave_std'])
```

## Moran Plot

```python
lisa = esda.Moran_Local(br['totCases'], w)
```

```python
# Break observations into significant or not
br['significant'] = lisa.p_sim < 0.05
# Store the quadrant they belong to
br['quadrant'] = lisa.q
```

```python
br['significant'].head()
```

```python
lisa.p_sim[:5]
```

```python
br['quadrant'].head()
```

```python
lisa_cluster(lisa, br);
```

```python
# Setup the figure and axis
fig, ax = plt.subplots(figsize=(20,6))
# Plot insignificant clusters
ns = br.loc[br['significant']==False, 'geometry']
ns.plot(ax=ax, color='k')
# Plot HH clusters
hh = br.loc[(br['quadrant']==1) & (br['significant']==True), 'geometry']
hh.plot(ax=ax, color='red')
# Plot LL clusters
ll = br.loc[(br['quadrant']==3) & (br['significant']==True), 'geometry']
ll.plot(ax=ax, color='blue')
# Plot LH clusters
lh = br.loc[(br['quadrant']==2) & (br['significant']==True), 'geometry']
lh.plot(ax=ax, color='#83cef4')
# Plot HL clusters
hl = br.loc[(br['quadrant']==4) & (br['significant']==True), 'geometry']
hl.plot(ax=ax, color='#e59696')
# Style and draw
f.suptitle('Local Indicators of Spatial Association (LISA) for covid 19 confirmed case', size=30)
f.set_facecolor('0.75')
ax.set_axis_off()
plt.show()
```
